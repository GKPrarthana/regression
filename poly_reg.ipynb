```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
X=dataset.iloc[:,1:2].values
y=dataset.iloc[:,2].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

```python
def viz_linear():
    plt.scatter(X,y,color='blue',label='Actual data')
    plt.plot(X,lin_reg.predict(X),color='red',label='Regression line')
    plt.title('Linear Regression model')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.legend()
    plt.show()
    return
```

```python
viz_linear()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)
pol_reg=LinearRegression()
pol_reg.fit(X_poly,y)
```

```python
def viz_polynomial():
    plt.scatter(X,y,color='blue',label='Actual data')
    plt.plot(X,pol_reg.predict(poly_reg.fit_transform(X)),color='red',label='Regression line')
    plt.title('Linear Regression with polynomial with degree 4')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.legend()
    plt.show()
    return
```

```python
viz_polynomial()
```
